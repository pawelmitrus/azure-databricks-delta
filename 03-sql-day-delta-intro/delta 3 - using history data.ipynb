### Disaster Recovery

```
%sql

INSERT OVERWRITE TABLE flights_parquet_delta 
SELECT * FROM flights_parquet_delta VERSION AS OF 0
```

```
%sql

--PREVIEW
RESTORE TABLE flights_parquet_delta TO VERSION AS OF 0
```

```
%sql

INSERT INTO flights_parquet_delta
  SELECT * FROM flights_parquet_delta VERSION AS OF 0
  WHERE date = 1234124
```

```
%sql

SELECT * FROM flights_parquet_delta
```

```
%sql

DESCRIBE HISTORY flights_parquet_delta
```

```
%sql

SET spark.databricks.delta.commitInfo.userMetadata=overwritten-for-fixing-incorrect-data;

INSERT OVERWRITE flights_parquet_delta SELECT * FROM flights_parquet_delta VERSION AS OF 1;

SET spark.databricks.delta.commitInfo.userMetadata=;
```

```
%sql
DESCRIBE HISTORY flights_parquet_delta
```
